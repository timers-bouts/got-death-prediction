```python
from __future__ import annotations
from pathlib import Path
import argparse
import json
import pandas as pd

def load_json(path: Path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)
```

```python

# 1. Use the helper function to load the JSON
episodes_raw = load_json(Path("episodes.json"))

# 2. Extract the nested list of episodes
episodes_raw = episodes_raw["episodes"]

# 3. Normalize the episodes list into a flat table
episodes = pd.json_normalize(episodes_raw, sep=".")

# 4. Extract and flatten the scenes
#    If scenes are nested inside each episode:
# scenes = pd.json_normalize(
#     episodes_raw,
#     record_path=["scenes"],
#     meta=["season", "episode", "title"],  # include episode info for context
#     sep="."
# )

episodes
```
